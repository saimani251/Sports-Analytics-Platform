```python
import pandas as pd
ball_dataset=pd.read_csv('ball_dataset.csv')
```

```python
ball_dataset.head(10)
```

```python
all_season=pd.read_csv('IPL_2008_to_2023.csv')
all_season.head(10)
```

```python
total_runs = ball_dataset.groupby(['match_id','innings']).sum()['TotalRuns'].add(1).reset_index()
total_runs = total_runs[total_runs['innings']==1]
total_runs
```

```python
match_data = all_season.merge(total_runs[['match_id','TotalRuns']],left_on='id',right_on='match_id')
match_data
```

```python
match_data['team1'] = match_data['team1'].replace('Delhi Daredevils','Delhi Capitals')
match_data['team2'] = match_data['team2'].replace('Delhi Daredevils','Delhi Capitals')

match_data['team1'] = match_data['team1'].replace('Deccan Chargers','Sunrisers Hyderabad')
match_data['team2'] = match_data['team2'].replace('Deccan Chargers','Sunrisers Hyderabad')

match_data['team1'] = match_data['team1'].replace('Kings XI Punjab','Punjab Kings')
match_data['team2'] = match_data['team2'].replace('Kings XI Punjab','Punjab Kings')
match_data['team1'].value_counts()
```

```python
teams = ['Kolkata Knight Riders', 
         'Chennai Super Kings',
         'Punjab Kings', 
         'Rajasthan Royals',
         'Mumbai Indians',
         'Delhi Capitals',
         'Royal Challengers Bangalore',
         'Sunrisers Hyderabad',
         'Lucknow Super Giants',
         'Gujarat Titans']
```

```python
match_data = match_data[match_data['team1'].isin(teams)]
match_data = match_data[match_data['team2'].isin(teams)]
```

```python
match_data['team1'].value_counts()
```

```python
match_data['team2'].value_counts()
```

```python
match_data
```

```python
match_data = match_data[['id','city','winner','TotalRuns']]
```

```python
match_data.rename(columns={'id':'match_id'},inplace=True)
```

```python
match_data.columns
```

```python
innings2 = match_data.merge(ball_dataset,on='match_id')
innings2
```

```python
innings2 = innings2[innings2['innings']==2]

innings2['batting_team'] = innings2['batting_team'].replace('Delhi Daredevils','Delhi Capitals')
innings2['bowling_team'] = innings2['bowling_team'].replace('Delhi Daredevils','Delhi Capitals')

innings2['batting_team'] = innings2['batting_team'].replace('Deccan Chargers','Sunrisers Hyderabad')
innings2['bowling_team'] = innings2['bowling_team'].replace('Deccan Chargers','Sunrisers Hyderabad')

innings2['batting_team'] = innings2['batting_team'].replace('Kings XI Punjab','Punjab Kings')
innings2['bowling_team'] = innings2['bowling_team'].replace('Kings XI Punjab','Punjab Kings')
```

```python
innings2 = innings2[innings2['batting_team'].isin(teams)]
innings2 = innings2[innings2['bowling_team'].isin(teams)]
print(innings2['TotalRuns_y'].dtype)
```

```python
innings2['curr_score'] = innings2.groupby('match_id')['TotalRuns_y'].cumsum()
innings2
```

```python
innings2['runs_left'] = innings2['TotalRuns_x'] - innings2['curr_score']
innings2
```

```python
innings2['balls_left'] = 126 - (innings2['Over']*6 +innings2['ball'])
innings2
```

```python
innings2['player_dismissed'] = innings2['player_dismissed'].fillna("0")
innings2['player_dismissed'] = innings2['player_dismissed'].apply(lambda x:x if x=="0" else "1")
innings2['player_dismissed'] = innings2['player_dismissed'].astype('int')
wickets = innings2.groupby('match_id')['player_dismissed'].cumsum()
innings2['wickets_remaining'] = 10-wickets
innings2
```

```python
innings2['curr_run_rate'] = round((innings2['curr_score']*6)/(120 - innings2['balls_left']),2)
innings2
```

```python
innings2['req_run_rate'] = round((innings2['runs_left']*6)/(innings2['balls_left']),2)
innings2
```

```python
def winner(team):
    return 1 if team['batting_team'] == team['winner'] else 0
```

```python
innings2['results'] = innings2.apply(winner,axis=1)
innings2
```

```python
feature_data = innings2[['batting_team','bowling_team','city','runs_left','balls_left','wickets_remaining','TotalRuns_x','curr_run_rate','req_run_rate','results']]
feature_data
```

```python
data = feature_data.sample(feature_data.shape[0])
data.dropna(inplace=True)
```

```python
data = data[data['balls_left']!=0]
```

```python
data.to_csv("data1.csv", index=False)
```

```python
from sklearn.model_selection import train_test_split as tts
x = data.iloc[:, :-1]
y = data.iloc[:,-1]
x_train,x_test,y_train,y_test = tts(x,y,test_size=0.20,random_state=100)
x_train
```

```python
x_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
],
remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

lr = Pipeline([
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
gradient_boosting_pipeline = Pipeline([
    ('preprocessor', trf),
    ('model', GradientBoostingClassifier())
])

random_forest_pipeline = Pipeline([
    ('preprocessor', trf),
    ('model', RandomForestClassifier(n_estimators=200,  
    max_depth=10,       
    min_samples_split=10,  
    min_samples_leaf=4,    
    max_features='sqrt',   
    random_state=0))
])

gradient_boosting_pipeline.fit(x_train, y_train)
random_forest_pipeline.fit(x_train, y_train)
gb_predictions = gradient_boosting_pipeline.predict(x_test)
rf_predictions = random_forest_pipeline.predict(x_test)
gb_accuracy = accuracy_score(y_test, gb_predictions)
rf_accuracy = accuracy_score(y_test, rf_predictions)

gb_auc = roc_auc_score(y_test, gradient_boosting_pipeline.predict_proba(x_test)[:, 1])
rf_auc = roc_auc_score(y_test, random_forest_pipeline.predict_proba(x_test)[:, 1])

gb_accuracy,rf_accuracy
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold

skf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)

lr_scores = cross_val_score(lr, x, y, cv=skf, scoring='accuracy')
gb_scores = cross_val_score(gradient_boosting_pipeline, x, y, cv=skf, scoring='accuracy')
rf_scores = cross_val_score(random_forest_pipeline, x, y, cv=skf, scoring='accuracy')

print("Logistic Regression CV Accuracy: %0.2f (+/- %0.2f)" % (lr_scores.mean(), lr_scores.std() * 2))
print("Gradient Boosting CV Accuracy: %0.2f (+/- %0.2f)" % (gb_scores.mean(), gb_scores.std() * 2))
print("Random Forest CV Accuracy: %0.2f (+/- %0.2f)" % (rf_scores.mean(), rf_scores.std() * 2))
```

```python
from sklearn.metrics import classification_report

print("Logistic Regression Report:\n", classification_report(y_test, y_pred))
print("Gradient Boosting Report:\n", classification_report(y_test, gb_predictions))
print("Random Forest Report:\n", classification_report(y_test, rf_predictions))

print("Gradient Boosting ROC AUC: %0.2f" % gb_auc)
print("Random Forest ROC AUC: %0.2f" % rf_auc)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def plot_confusion_matrix(cm, classifier_name):
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {classifier_name}')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()


cm_lr = confusion_matrix(y_test, y_pred)
cm_gb = confusion_matrix(y_test, gb_predictions)
cm_rf = confusion_matrix(y_test, rf_predictions)

plot_confusion_matrix(cm_lr, 'Logistic Regression')
plot_confusion_matrix(cm_gb, 'Gradient Boosting')
plot_confusion_matrix(cm_rf, 'Random Forest')
```
